```python
import pandas as pd
#diccionario simple pero con datos sucios
data = {"Id_producto": [1011,1002,1003,1003],
        "Cantidad_vendida":[30,None,25,25],
        "Precio":[20.5,15.5,None,22.5]
        }
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
#para saber numero de valores faltantes por columna
df.isnull().sum()
```

```python
#eliminar valores no habilitados
df_eliminados = df.dropna()
df_eliminados
#la desventaja es que se borra siempre la fila o registro completo
```

```python
#primero rellena un valor faltante con un valor específico
#posteriormente rellena un valor faltante con el promedio de su columna
valores_nuevos = {"Cantidad_vendida": 0, "Precio":df["Precio"].mean()}
#fillna rellena los valores faltantes con el valor que se le indique
df_rellenados = df.fillna(valores_nuevos)
df_rellenados
```

```python
#cambiar el tipo de datos de una serie
df_rellenados["Cantidad_vendida"] = df_rellenados["Cantidad_vendida"].astype(int)
df_rellenados
```
